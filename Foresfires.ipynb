```python
import numpy as np
import pandas as pd
data = pd.read_csv('forestfires.csv')
```

```python
Data_an = pd.DataFrame(columns = ('X', 'Y', 'sum', 'month', 'day', 'FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area'))
Di = 0
for x in range(1, 10):
    for y in range(2, 10):
        value = 0
        for i, vl in data.iterrows():
            if (vl['X'] == x) & (vl['Y'] == y):
                value += 1
        if value != 0:
            print("{} {} = {}".format(x, y, value))
            data_now = data[(data['X'] == x) & (data['Y'] == y)]
            data_now_mn = data_now.mean()
            Data_an.loc[Di] = [x, y, value, data_now['month'].value_counts().index[0], data_now['day'].value_counts().index[0], data_now_mn['FFMC'], data_now_mn['DMC'], data_now_mn['DC'], data_now_mn['ISI'], data_now_mn['temp'], data_now_mn['RH'], data_now_mn['wind'], data_now_mn['rain'], data_now_mn['area']]
            Di += 1
Data_an = Data_an.sort_values(by = ['sum'], ascending = False).reset_index(drop = True)
Data_an
```

```python
Data_PCA = Data_an[:4]
Data_PCA
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit(data.day)
data.day = label.transform(data.day)
month = pd.get_dummies(data['month']) 
data = pd.concat((data, month), axis=1)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler 
scale_features_std = StandardScaler() 
features = scale_features_std.fit_transform(data[['rain', 'area']])
```

```python
from sklearn.preprocessing import MinMaxScaler 
scale_features_mm = MinMaxScaler() 
data[['rain', 'area']] = scale_features_mm.fit_transform(data[['rain', 'area']]) 
data
```

```python
for x in Data_PCA['X']:
    for y in Data_PCA[Data_PCA['X'] == x]['Y']:
        if (x == 3) & (y == 4):
            print('\n For X = 3, Y = 4 and X = 4, Y = 4:')
            data_now = data[((data['X'] == 3) | (data['X'] == 4)) & (data['Y'] == 4)]
        else:
            print('\n For X = {}, Y = {}:'.format(x, y))
            data_now = data[(data['X'] == x) & (data['Y'] == y)]
        target = data_now.temp
        train = data_now.drop(['X', 'Y', 'month'], axis=1)
        from sklearn.model_selection import train_test_split
        X_train, X_test, y_train, y_test = train_test_split(train, target, test_size = 0.3, random_state = 42)

        N_train, _ = X_train.shape 
        N_test,  _ = X_test.shape 

        from sklearn.decomposition import PCA
        %matplotlib inline
        import matplotlib.pyplot as plt
        pca = PCA()
        pca.fit(X_train)
        X_pca = pca.transform(X_train)
        for i, component in enumerate(pca.components_):
            print("{} component: {}% of initial variance".format(i + 1, 
                  round(100 * pca.explained_variance_ratio_[i], 2)))
            print(" + ".join("%.3f x %s" % (value, name)
                             for value, name in zip(component,train.columns)))
        print('\n')
```

```python
target = data.temp
train = data.drop(['month'], axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, target, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

from sklearn.decomposition import PCA
%matplotlib inline
import matplotlib.pyplot as plt
pca = PCA()
pca.fit(X_train)
X_pca = pca.transform(X_train)
for i, component in enumerate(pca.components_):
    print("{} component: {}% of initial variance".format(i + 1, 
          round(100 * pca.explained_variance_ratio_[i], 2)))
    print(" + ".join("%.3f x %s" % (value, name)
                     for value, name in zip(component,train.columns)))
```

```python
plt.figure(figsize=(10,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)
plt.axhline(0.9, c='r')
plt.axvline(3, c='b')
```
